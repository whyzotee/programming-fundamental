Exercise 5.1 ให้เขียน function ชื่อ day_of_year(day, month ,year) โดยมีการคืนค่า คือ day_of_years เป็นวันที่ลำดับที่เท่าใดของปีคริสตศักราช year
ปีที่เป็น Leap Year เดือนกุมภาพันธ์จะมี 29 วัน
ให้สร้างฟังก์ชัน is_leap เพื่อตรวจสอบ leap year แยกออกมา และให้ฟังก์ชัน day_of_year เรียกใช้ is_leap อีกที


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def leap_year(year):
  year_str = str(year)
  year_only_first = year_str[0]+ "0"*(len(year_str[1:]))

  if year % 4 == 0 and year % 100 == 0 and year % 400 == 0:
    return True
  elif year % 4 == 0 and year % 100 != 0 and int(year_only_first) % 400 == 0:
    return True
  else:
    return False


def day_of_year(day, month, year):
  is_leap_year = leap_year(year)

  all_days = day

  if month > 1:
    all_days += sum(days_in_month[:month - 1])

  if is_leap_year and month > 2:
      all_days += 1

  return all_days

# Testcases
print("Normal year (2023):")
print(f"1 Jan 2023: {day_of_year(1, 1, 2023)}")  # ควรได้ 1
print(f"1 Feb 2023: {day_of_year(1, 2, 2023)}")  # ควรได้ 32
print(f"1 Mar 2023: {day_of_year(1, 3, 2023)}")  # ควรได้ 60
print(f"31 Dec 2023: {day_of_year(31, 12, 2023)}")  # ควรได้ 365

print("\nLeap year (2024):")
print(f"1 Jan 2024: {day_of_year(1, 1, 2024)}")  # ควรได้ 1
print(f"29 Feb 2024: {day_of_year(29, 2, 2024)}")  # ควรได้ 60
print(f"1 Mar 2024: {day_of_year(1, 3, 2024)}")  # ควรได้ 61
print(f"31 Dec 2024: {day_of_year(31, 12, 2024)}")  # ควรได้ 366

print("\nCentury year, not a leap year (1900):")
print(f"1 Mar 1900: {day_of_year(1, 3, 1900)}")  # ควรได้ 60

print("\nCentury year, leap year (2000):")
print(f"1 Mar 2000: {day_of_year(1, 3, 2000)}")  # ควรได้ 61

print("\nMid-year dates:")
print(f"15 Jun 2023: {day_of_year(15, 6, 2023)}")  # ควรได้ 166
print(f"15 Jun 2024: {day_of_year(15, 6, 2024)}")  # ควรได้ 167 (Leap year)

print("\nLast day of each quarter:")
print(f"31 Mar 2023: {day_of_year(31, 3, 2023)}")  # ควรได้ 90
print(f"30 Jun 2023: {day_of_year(30, 6, 2023)}")  # ควรได้ 181
print(f"30 Sep 2023: {day_of_year(30, 9, 2023)}")  # ควรได้ 273
print(f"31 Dec 2023: {day_of_year(31, 12, 2023)}")  # ควรได้ 365
````

Exercise 5.2 ให้เขียนฟังก์ชัน เพิ่มเติมเป็น date_diff
รับข้อมูลในรูปแบบ “dd-mm-yyyy” เช่น date_diff(“1-1-2018”, “1-1-2020”) จะได้ 731 วัน date_diff(“25-12-1999”, “9-3-2000”) จะได้ 76 วัน
ให้เขียนฟังก์ชัน day_in_year โดยจะส่งค่าจำนวนวันของปี (365 หรือ 366) โดยรับข้อมูลเป็น ปี
ส่งคืนข้อมูลเป็นจำนวนวันตั้งแต่วันที่แรก จนถึงวันที่สอง โดยรวมทั้ง 2 วันนั้นเข้าไปด้วย
ให้สมมติว่าวันแรก จะต้องมาก่อนวันที่สองเสมอ ดังนั้นไม่ต้องตรวจสอบ


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def leap_year(year):
  year_str = str(year)
  year_only_first = year_str[0]+ "0"*(len(year_str[1:]))

  if year % 4 == 0 and year % 100 == 0 and year % 400 == 0:
    return True
  elif year % 4 == 0 and year % 100 != 0 and int(year_only_first) % 400 == 0:
    return True
  else:
    return False

def day_of_year_from_start(day, month, year):
  is_leap_year = leap_year(year)

  sum_of_days = 0

  if day == 1 and month == 1:
    sum_of_days += sum(days_in_month)
  else:
    sum_of_days += (days_in_month[month - 1] - day) + 1
    sum_of_days += sum(days_in_month[month:])

  if is_leap_year and month <= 2:
      sum_of_days += 1

  return sum_of_days

def day_of_year_from_end(day, month, year):
  is_leap_year = leap_year(year)

  sum_of_days = sum(days_in_month[:month - 1]) + day

  if is_leap_year and month > 2:
      sum_of_days += 1

  return sum_of_days

def calculate_n_year(start_year,year_count):
  sum = 0
  for i in range(1, year_count):
    sum += day_of_year_from_start(1, 1, start_year+i)
  return sum

def split_date(date, format='-'):
  return [int(n) for n in date.split(format)]

def date_diff(date1, date2):
  dayA, monthA, yearA = split_date(date1)
  dayB, monthB, yearB = split_date(date2)

  diff_days = 0
  all_days_start_year = day_of_year_from_start(dayA, monthA, yearA)
  all_days_end_year = day_of_year_from_end(dayB, monthB, yearB)

  if yearA == yearB:
    diff_days = all_days_end_year
  else:
    if yearB - yearA > 1:
      diff_days += calculate_n_year(yearA, yearB - yearA)
    diff_days += (all_days_start_year + all_days_end_year)

  return diff_days


# ทดสอบฟังก์ชัน
print(date_diff("1-1-2018", "1-1-2020"))  # ควรได้ 731
print(date_diff("25-12-1999", "9-3-2000"))  # ควรได้ 76
print(date_diff("1-1-2000", "1-1-2001"))  # ควรได้ 367 (ปี 2000 เป็น leap year)
print(date_diff("1-1-2001", "1-1-2002"))  # ควรได้ 366 (ปี 2001 ไม่ใช่ leap year)
print(date_diff("1-3-2000", "1-3-2001"))  # ควรได้ 366 (ข้ามปี leap year)
print(date_diff("1-1-2023", "31-12-2023"))  # ควรได้ 365 (ทั้งปี 2023)
````

Exercise 5.3 จากโปรแกรม 5.2 ให้เขียนฟังก์ชัน date_diff  เพิ่มเติม โดยให้มีการตรวจสอบ
วันที่ต้องเป็นวันที่ถูกต้องของเดือนนั้นๆ
เดือนต้องอยู่ระหว่าง 1-12
เดือนกุมภาพันธ์ของปีที่มี Leap Year เท่านั้นที่จะมี 29 วันได้
หากข้อมูล Input ผิดพลาด ให้ Return -1


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def leap_year(year):
  year_str = str(year)
  year_only_first = year_str[0]+ "0"*(len(year_str[1:]))

  if year % 4 == 0 and year % 100 == 0 and year % 400 == 0:
    return True
  elif year % 4 == 0 and year % 100 != 0 and int(year_only_first) % 400 == 0:
    return True
  else:
    return False

def day_of_year_from_start(day, month, year):
  is_leap_year = leap_year(year)

  sum_of_days = 0

  if day == 1 and month == 1:
    sum_of_days += sum(days_in_month)
  else:
    sum_of_days += (days_in_month[month - 1] - day) + 1
    sum_of_days += sum(days_in_month[month:])

  if is_leap_year and month <= 2:
      sum_of_days += 1

  return sum_of_days

def day_of_year_from_end(day, month, year):
  is_leap_year = leap_year(year)

  sum_of_days = sum(days_in_month[:month - 1]) + day

  if is_leap_year and month > 2:
      sum_of_days += 1

  return sum_of_days

def calculate_n_year(start_year,year_count):
  sum = 0
  for i in range(1, year_count):
    sum += day_of_year_from_start(1, 1, start_year+i)
  return sum

def split_date(date, format='-'):
  list_date = [n for n in date.split(format)]
  error = -1

  if len(list_date) != 3:
    return error

  for i in range(len(list_date)):
    if not f"{list_date[i]}".isdigit():
      return error
    else:
      list_date[i] = int(list_date[i])

  day = list_date[0]
  month = list_date[1]
  year = list_date[2]

  if (1 < day > 31) or (0 < month > 12) or year < 1:
    return error
  elif day > days_in_month[month - 1]:
    return error
  else:
    return list_date

def date_diff(date1, date2):
  format_dateA, format_dateB = split_date(date1), split_date(date2)

  if format_dateA == -1 or format_dateB == -1:
    return -1

  dayA, monthA, yearA = format_dateA
  dayB, monthB, yearB = format_dateB

  diff_days = 0
  all_days_start_year = day_of_year_from_start(dayA, monthA, yearA)
  all_days_end_year = day_of_year_from_end(dayB, monthB, yearB)

  if yearA == yearB:
    diff_days = all_days_end_year
  else:
    if yearB - yearA > 1:
      diff_days += calculate_n_year(yearA, yearB - yearA)
    diff_days += (all_days_start_year + all_days_end_year)

  return diff_days

# ทดสอบฟังก์ชัน
print(date_diff("1-1-2018", "1-1-2020"))  # ควรได้ 731
print(date_diff("25-12-1999", "9-3-2000"))  # ควรได้ 76
print(date_diff("1-1-2000", "1-1-2001"))  # ควรได้ 367 (ปี 2000 เป็น leap year)
print(date_diff("1-1-2001", "1-1-2002"))  # ควรได้ 366 (ปี 2001 ไม่ใช่ leap year)
print(date_diff("1-3-2000", "1-3-2001"))  # ควรได้ 366 (ข้ามปี leap year)
print(date_diff("1-1-2023", "31-12-2023"))  # ควรได้ 365 (ทั้งปี 2023)

# ทดสอบกรณีข้อมูลไม่ถูกต้อง
print(date_diff("31-4-2023", "1-5-2023"))  # ควรได้ -1 (เดือนเมษายนมี 30 วัน)
print(date_diff("29-2-2023", "1-3-2023"))  # ควรได้ -1 (ปี 2023 ไม่ใช่ leap year)
print(date_diff("1-13-2023", "1-1-2024"))  # ควรได้ -1 (เดือนไม่ถูกต้อง)
print(date_diff("abc", "1-1-2024"))  # ควรได้ -1 (รูปแบบวันที่ไม่ถูกต้อง)
````

2.	ให้เขียน function find_mismatch ซึ่งรับ string มา 2 ค่า โดยจะส่งค่ากลับดังนี้
- 0 เมื่อ string ทั้ง 2 ตัว เหมือนกัน โดยไม่สนใจ case
- 1 เมื่อ string 1 จะเหมือน string 2 โดยการลบและ/หรือเพิ่มตัวอักษรเพียง 1 ตัว
- 2 เมื่อ string ไม่เข้าเกณฑ์ 0 หรือ 1


| String  1| 	String2	|ผลลัพธ์|
|----------|----------|------|
|Python|	Java	|2|
|book	 |  boot	|1|
|fan	 |  fun	  |1|
|sin	 | sink	  |1 เพราะเพิ่ม k ใน str2 แล้วจะเหมือนกัน
|Dog	 | dog	  |0
|poke	 | spoke	|1 เพิ่ม s ใน str2 แล้วเหมือนกัน
|poker |	poke	|1 ตัด r ใน str1 ออกแล้วเหมือนกัน






````python
def count_same_char(word, diff_word):
  sum = 0
  checked_word = []
  for i in word:
    if i in diff_word and i not in checked_word:
      if word.count(i) > diff_word.count(i):
        sum += diff_word.count(i)
      else:
        sum += word.count(i)

      checked_word.append(i)

  return sum

def find_word_next(str1, str2):
  if str1[1:] == str2:
    return str1[0], "remove"
  if str1[0:-1] == str2:
    return str1[-1], "remove"
  if str2[1:] == str1:
    return str2[0], "add"
  if str2[0:-1] == str1:
    return str2[-1], "add"


def find_diff_word(str1, str2):
  s1, s2 = str1.lower(), str2.lower()
  lenS1, lenS2 = len(s1), len(s2)

  if abs(lenS1 - lenS2) == 1:
    result = find_word_next(s1, s2)
    if result != None:
      return result

  if (s1[:-1] == s2[:-1]) or (s1[1:] == s2[1:]):
    return "x", ""

  same_char_str1 = count_same_char(s1, s2)
  same_char_str2 = count_same_char(s2, s1)

  if abs(lenS1 - same_char_str1) == 1 and abs(lenS2 - same_char_str2) == 1:
    return "x", ""

  return "", ""

def find_mismatch(str1, str2):
  if str1.lower() == str2.lower():
    return 0

  diff, type_of_diff = find_diff_word(str1, str2)
  if len(diff) == 1:
    if type_of_diff == '':
      return 1
    return f'1 {type_of_diff} {diff}'

  return 2


wordA = ["Python", "book", "fan", "sin", "Dog", "poke", "poker"]
wordB = ["Java", "boot", "fun", "sink", "dog", "spoke", "poke"]

print("-"*35)
print("| String 1 ", " String2 ", '    ผลลัพธ์   |', sep='|')
print("-"*35)

for i in range(len(wordA)):
  result = find_mismatch(wordA[i], wordB[i])
  print(f'| {wordA[i]:<8} | {wordB[i]:<7} | {result:<10} |')
  print("-"*35)
````

````python
def count_same_char(word, diff_word):
  sum = 0
  checked_word = []
  for i in word:
    if i in diff_word and i not in checked_word:
      if word.count(i) > diff_word.count(i):
        sum += diff_word.count(i)
      else:
        sum += word.count(i)

      checked_word.append(i)

  return sum

def find_word_next(str1, str2):
  if str1[1:] == str2:
    return str1[0], "remove"
  if str1[0:-1] == str2:
    return str1[-1], "remove"
  if str2[1:] == str1:
    return str2[0], "add"
  if str2[0:-1] == str1:
    return str2[-1], "add"


def find_diff_word(str1, str2):
  s1, s2 = str1.lower(), str2.lower()
  lenS1, lenS2 = len(s1), len(s2)

  if abs(lenS1 - lenS2) == 1:
    result = find_word_next(s1, s2)
    if result != None:
      return result

  if (s1[:-1] == s2[:-1]) or (s1[1:] == s2[1:]):
    return "x", ""

  same_char_str1 = count_same_char(s1, s2)
  same_char_str2 = count_same_char(s2, s1)

  if abs(lenS1 - same_char_str1) == 1 and abs(lenS2 - same_char_str2) == 1:
    return "x", ""

  return "", ""

def find_mismatch(str1, str2):
  if str1.lower() == str2.lower():
    return 0

  diff, type_of_diff = find_diff_word(str1, str2)
  if len(diff) == 1:
    return 1

  return 2

def check_and_format(input):
  is_eng = "abcdefghijklmnopqrstuvwxyz"
  list_word = input.split()

  for i in range(len(list_word)):
    for j in list_word[i].lower():
      if j not in is_eng:
        return -1
    list_word[i] = list_word[i].lower()

  return list_word

def spelling_correction(input_string, correct_spells):
  words = check_and_format(input_string)

  if words == -1:
    return "Error"

  output = ""

  for word in words:
    diff_one = []
    for correct_word in correct_spells:
      result = find_mismatch(word, correct_word)
      if result == 1:
        diff_one.append(correct_word)

    if len(diff_one) > 0:
      output += diff_one[-1] + " "
    else:
      output += word + " "

  return output.strip()


word = ["Thes is the Firs cas", "programming is fan and easy", "Thes is vary essy", "Wee ipve Pythen"]
correct_spells = [
  ['that', 'first', 'case', 'car'],
  ['programming', 'this','fun', 'easy', 'book'],
  ['this', 'is', 'very', 'easy'],
  ['we', 'live', 'ln', 'Python']
]

space = 112

print("-"*space)
print(f"| {'ประโยค':^30} ", f"{'correct_spells':^48}", f" {'output string':^27} |", sep='|')
print("-"*space)

for i in range(len(word)):
  result = spelling_correction(word[i], correct_spells[i])
  print(f'| {word[i]:<30} | {f"{correct_spells[i]}":<46} | {result:<27} |')
  print("-"*space)
````

````python
def find_word_horizontal(crosswords, word):
    for i in range(len(crosswords)):
        for j in range(len(crosswords[i]) - (len(word) - 1)):
            n_word = ""
            for k in range(len(word)):
              n_word += crosswords[i][j+k]
            if n_word == word:
                return [i, j]
    return None

crosswords = [
    ['s', 'd', 'o', 'g'],
    ['c', 'u', 'c', 'm'],
    ['a', 'c', 'a', 't'],
    ['t', ' e', 't', 'k']
]

word = 'cat'

result = find_word_horizontal(crosswords, word)
print("Output",result)
````
